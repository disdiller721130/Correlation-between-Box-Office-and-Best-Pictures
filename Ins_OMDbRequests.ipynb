```python
def month_convert(mon):
    abbr_to_num = {name: num for num, name in enumerate(calendar.month_abbr) if num}
    return(abbr_to_num[mon])

def datetime_convert(year,month,date):
    dateTime = datetime.date(year,month,date)
    unixtime = time.mktime(dateTime.timetuple())
    return(unixtime)
```
