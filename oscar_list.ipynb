```python
import requests
import json
from pprint import pprint
import pandas as pd
import numpy as np
api_key="hwXdsTMehMpXGUVpUK"
```

```python
oscar_movie="Book1.csv"
oscar_movie_list=pd.read_csv(oscar_movie)
oscar_movie_list["Director"]=""
oscar_movie_list["Metascore"]=""
oscar_movie_list["Box Office"]=""
oscar_movie_list["IMDB Vote"]=""
oscar_movie_list["Genre"]=""
oscar_movie_list["Actors"]=""
```

```python
for index,row in oscar_movie_list.iterrows():
    genre_list=[]
    try:
        movie_name=row["Best picture list"]
        url = "http://www.omdbapi.com/?t="
        api_key = "&apikey=trilogy"
        response = requests.get(url + movie_name + api_key).json()
        oscar_movie_list.loc[index,"Director"]=response['Director']
        oscar_movie_list.loc[index,"Box Office"]=response['BoxOffice']
        oscar_movie_list.loc[index,"Metascore"]=response['Metascore']
        oscar_movie_list.loc[index,"IMDB Vote"]=response['imdbVotes']
        oscar_movie_list.loc[index,"Actors"]=response['Actors']
        genre_list.append(response['Genre'])
        movie_genre=genre_list[0].split(',')
        oscar_movie_list.loc[index,"Genre"]=movie_genre[0]
    except KeyError:
        oscar_movie_list.loc[index,"Box Office"]=np.NaN
    
oscar_movie_list=oscar_movie_list.dropna(how='any')
oscar_movie_list
```

```python
movie_list_refine=oscar_movie_list.loc[oscar_movie_list["Box Office"]!="N/A",:]
movie_list_refine["Box Office"]=movie_list_refine["Box Office"].replace({'\$':''}, regex = True)
movie_list_refine
```
